```python
import pandas as pd

# Load the transcriptomics data
transcriptomics_data = pd.read_csv('/Users/yahyafarooqi/Documents/AntiGEM/Data/intermediate/flux/nim2.chip21-batch26.180402.csv')
transcriptomics_data.to_csv('/Users/yahyafarooqi/Documents/AntiGEM/Data/intermediate/transcriptomics_data.tsv', sep='\t')

# Read the saved tsv file
df = pd.read_csv('/Users/yahyafarooqi/Documents/AntiGEM/Data/intermediate/transcriptomics_data.tsv', sep='\t')

# Select the relevant columns and rename them
df = df.iloc[:, [1, 2]]  # Select only the 2nd and 3rd columns
df.columns = ['gene', 'value']  # Rename columns to 'gene' and 'value'

# Convert 'gene' to string and 'value' to numeric
df['gene'] = df['gene'].astype(str)
df['value'] = pd.to_numeric(df['value'], errors='coerce')  # Converts non-numeric values to NaN if present

# Load gene_df from a TSV file
gene_df = pd.read_csv('/Users/yahyafarooqi/Documents/AntiGEM/Data/intermediate/gene_df2.tsv', sep='\t')

# Filter gene_df by the gene names in df
filtered_gene_df = gene_df[gene_df['gene'].isin(df['gene'])]

# Merge the filtered_gene_df with df
merged_data = pd.merge(filtered_gene_df, df, on='gene', how='inner')
#drop gene column
merged_data = merged_data.drop(columns=['gene'])

# Save the merged dataframe
merged_data.to_csv('/Users/yahyafarooqi/Documents/AntiGEM/Data/intermediate/merged_data.tsv', sep='\t', index=False)

# Display the merged dataframe
print(merged_data)
```

```python
import cobra

model = cobra.io.read_sbml_model('/Users/yahyafarooqi/Documents/AntiGEM/models/subtilis_ iYO844.xml')

```

```python
model.genes.
```

```python
data = []

for gene in model.genes:
    gene_id = gene.id
    gene_name = gene.name
    if gene_id and gene_name:  # Check if gene_id and gene_name are not None or empty
        # Append the gene_id and gene_name to the data list as a dictionary
        data.append({'gene_id': gene_id, 'gene': gene_name})

gene_df = pd.DataFrame(data, columns=['gene_id', 'gene'])

print(gene_df.head())

#save to dtaframe

gene_df.to_csv('/Users/yahyafarooqi/Documents/AntiGEM/Data/intermediate/gene_df2.csv', index=False)
```
